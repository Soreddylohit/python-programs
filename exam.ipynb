# 1) write a function "prime" which will give output like the following.

```python
def prime(a):
    for i in range(2,a):
        if a%i==0:
            print("not_prime" )
            return
        else:
            print("prime")
            return
```

```python
prime(23)
```

```python
prime(20)
```

# If a word, phrase, or sequence that reads the same backwards as forward  then we call it a palindrome. for example "abba" and "abcba" are palindromes 

# write a function that will detect palindromes like the function

```python
def palindrome(str):
    if(str==str[::-1]):
        print("palindrome")
    else:
        print("not_palindrome")
```

```python
palindrome("abba")
```

```python
palindrome("abcba")
```

```python
palindrome("abcbav")
```

# LCM: the smallest number that is a multiple of each of two  numbers

# implement LCM using lambda function

```python
def LCM(x,y):
    if x>y:
        greater=x
    else:
        greater=y
    while(True):
        if((greater%x==0) and (greater%y==0)):
            LCM=greater
            break
        greater+=1
    return LCM
        
```

```python
LCM(10,12)
```

```python
LCM(10,13)
```
